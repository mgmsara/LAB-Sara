```python
#Filter the data in state and gender column to standardize the texts in those columns
cars_data_filtered['Gender'].value_counts()
```

```python
cars_data_filtered['State'].value_counts()
```

```python

def clean_state(x):
    if x in ['Cali', 'California']:
        return 'California'
    if x in ['WA', 'Washington']:
        return 'Washington'
    elif x in ['Arizona', 'AZ']:
        return 'Arizona'
    elif np.nan:  pass
    else:
        return 'U'

cars_data_filtered['State']= cars_data_filtered['State'].apply(clean_state)
```

```python
cars_data_filtered['State'].value_counts()
```

```python
cars_data_filtered
```

```python
cars_data_filtered['Gender'].value_counts()
```

```python
def clean_gender(x):
    if x in ['M','Male', "male"]:
        return 'Male'
    elif x in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
cars_data_filtered['Gender']= cars_data_filtered['Gender'].apply(clean_gender)
```
