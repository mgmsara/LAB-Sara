### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
## Your Code here
#LOC
vehicles_df[vehicles_df["Make"].str.contains("Dutton")] 
```

```python
#students_df.iloc[:4, 4] = [40, 50, 60, 70]

vehicles_df.loc[vehicles_df["Make"].str.contains("Dutton"), 'Make'] = 'Dutton'
```

```python
vehicles_df[vehicles_df["Make"].str.contains("Dutton")] 
```

```python
#Convert CO2 Emissions from Grams/Mile to Grams/Km
#1 Mile = 1.60934 Km
#students_df["py-score"]=students_df["py-score"]  * 1.5

vehicles_df["CO2 Emission Grams/Mile"] = vehicles_df['CO2 Emission Grams/Mile'] / 1.6 

vehicles_df["CO2 Emission Grams/Mile"]

```

```python
vehicles_df.rename(columns={"CO2 Emission Grams/Mile": "CO2 Emission Grams/km"})

```

```python
#Create a binary column that solely indicates if the transmission of a car is automatic or manual. 
#Use pandas.Series.str.startswith and convert MPG columns to km_per_liter 



def auto_manual_function(Transmission):
    if Transmission.startswith('A'):
        return "Auto"
    else:
        return "Manual"
    

#vehicles_df["AUTO_or_MANUAL"]= vehicles_df["AUTO_or_MANUAL"].apply(auto_manual_function)


vehicles_df.insert(loc = 15, column ='AUTO_or_MANUAL', value = vehicles_df["Transmission"].apply(auto_manual_function))

vehicles_df

```

```python
#Use pandas.Series.str.startswith and convert MPG columns to km_per_liter 
#Converting Gallons to Liters
#1 Gallon = 3.78541 Liters

def update_km_per_liter(x):
    return x / 1.6 * 3.7 
    

vehicles_df[["City MPG", 'Highway MPG', 'Combined MPG']]= vehicles_df[['City MPG', 'Highway MPG', 'Combined MPG']].apply(update_km_per_liter)

vehicles_df.rename(columns={"City MPG": "City KMPL"})


vehicles_df
```
